```python
# Introduction to seaborn # seaborn is very powerfull visualizatio(STATISTIC VISULAIZATION) pkg in python

import matplotlib.pyplot as plt # visulaiztion
import seaborn as sns # distribution visualtion 
# seaborn are used for advance visualization e.x --> distribution plot, line plot

%matplotlib inline 
plt.rcParams['figure.figsize'] = 9,5

import warnings
warnings.filterwarnings('ignore') # os error
```

```python
data.head() 
```

```python
data["InternetUsers"]
```

```python
# Distributions: 
vis1 = sns.distplot(data["InternetUsers"])  # UNIVARIATE ANALYSIS - statistics 
plt.show(vis1)
```

```python
vis1 = sns.distplot(data["InternetUsers"], bins=10) 
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(
    data=data, 
    x="IncomeGroup", 
    y="BirthRate", 
    palette="coolwarm", 
    hue="IncomeGroup", 
    dodge=False
)
plt.title("Birth Rate by Income Group", fontsize=14)
plt.xlabel("Income Group", fontsize=12)
plt.ylabel("Birth Rate", fontsize=12)
plt.show()
```

```python
# visualizing with seaborn
```

```python
vis4 = sns.lmplot(data = data,x = 'InternetUsers', y = 'BirthRate')
plt.show(vis4)
```

```python
vis3 = sns.lmplot(data = data,x = 'InternetUsers', y = 'BirthRate', fit_reg = True ) #lm - linear model
plt.show(vis3)
```

```python
vis31 = sns.lmplot(data = data,x = 'InternetUsers', y = 'BirthRate', fit_reg = False) #lm - linear model
plt.show(vis31)
```

```python
vis5 = sns.lmplot(data = data,x = 'InternetUsers', y = 'BirthRate',
                  fit_reg = False,hue = 'IncomeGroup') #hue - parameter for color
plt.show(vis5)
```

```python
vis5 = sns.lmplot(data=data, x='InternetUsers', y='BirthRate',
                  fit_reg=True, hue='IncomeGroup', height=7)
plt.show()
```
