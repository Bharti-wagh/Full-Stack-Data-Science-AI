using python split() function -- convert address, date, time. (Create more variable) 

```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\Bharti\Downloads\TASK_clean data.xlsx") 
```

```python
df
```

```python
df[['City', 'State', 'Country']] = df['ADDR'].str.split(', ', expand=True)
```

```python
df
```

```python
df[['Year', 'Month', 'Day']] = df['DT'].astype(str).str.split('-', expand=True)
```

```python
df
```

```python
df[['Hour', 'Minute','Second']] = df['TIME'].astype(str).str.split(':', expand=True)
```
