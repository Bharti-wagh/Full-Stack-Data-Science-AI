# python patterns.

1.Right Angle Triangle patterns.

```python
for i in range(1,6):
    print(" * " * i)
```

2.Inverted Right Angle Triangle patterns.

```python
for i in range(5,0,-1):
    print(" * " *i)
```

3.pyramid patterns

```python
for i in range(1,6):
    print(''*(5-i)+' * '*(2*i-1))
```

4.Inverted Pyramid Pattern.

```python
for i in range(5,0,-1):
    print(''*(5-i)+' * '*(2*i-1))
```

5.Dimond Pattern.

```python
for i in range(1,6):
    print(''*(5-i)+' * '*(2*i-1))
for i in range(4,0,-1):
    print(''*(5-i)+' * '*(2*i-1))
```

6.Hallow Square Pattern

```python
for i in range(5):
    for j in range(5):
        if i==0 or i==4 or j==0 or j==4:
            print('*',end='')
        else:
            print(' ',end='')
    print()
```

7.Full quare patterns.

```python
for i in range(5):
    print(' * '*5)
```

8.Right Angle Triangle patterns.Number pattern>

```python
for i in range(1,6):
    print(' '.join(str(x) for x in range(1,i+1)))
```

9.Inverted Right Angle Triangle.

```python
for i in range(5,0,-1):
    print(' '.join(str(x) for x in range(1,i+1)))
```

10.Floyds Triangle.

```python
num=1
for i in range(1,6):
    for j in range(1,i+1):
        print(num,end=' ')
        num+=1
    print()
```

11.Hallow Right Angle Triangle.

```python
n = 5
for i in range(1, n + 1):
    for j in range(1, i + 1):
        if j == 1 or j == i or i == n:
            print('*', end=' ')
        else:
            print(' ', end=' ')
    print()
```

12.Hallow pyramid pattern

```python
for i in range(1,6):
    
    for j in range(5-i):
        print(' ',end=' ')
    for j in range(2*i-1):
            if j == 0 or j == 2*i-2 or i == 5:
                print('*',end=' ')
            else:
                print(' ', end=' ')
    print()
```

13. Hallow Dimond Pattern.

```python
n = 5
for i in range(1, n + 1):
    for j in range(n - i):
        print(' ', end=' ')
    for j in range(1, 2 * i):
        if j == 1 or j == 2 * i - 1:
            print(i, end=' ')
        else:
            print(' ', end=' ')
    print()

for i in range(n - 1, 0, -1):
    for j in range(n - i):
        print(' ', end=' ')
    for j in range(1, 2 * i):
        if j == 1 or j == 2 * i - 1:
            print(i, end=' ')
        else:
            print(' ', end=' ')
    print()
```

14.Butterfly Pattern.

```python
n = 5
for i in range(1, n + 1):
    for j in range(i):
        print("*", end=" ")
    for j in range(2 * (n - i)):
        print(" ", end=" ")
    for j in range(i):
        print("*", end=" ")
    print()
for i in range(n, 0, -1):
    for j in range(i):
        print("*", end=" ")
    for j in range(2 * (n - i)):
        print(" ", end=" ")
    for j in range(i):
        print("*", end=" ")
    print()
```

15.Hallow number pyramid

```python
n = 5

for i in range(1, n + 1):
    
    for j in range(n - i):
        print(' ', end=' ')
    for j in range(1, 2 * i):
        if j == 1 or j == 2 * i - 1 or i == n:
            print(i, end=' ')
        else:
            print(' ', end=' ')
    print()
```

16.Full Star Pyramid

```python
n = 5

for i in range(1, n + 1):
    
    for j in range(n - i):
        print(' ', end=' ')
    
    for j in range(2 * i - 1):
        print('*', end=' ')
    print()
```

17.inverted full star pyramid

```python
n = 5

for i in range(n, 0, -1):
    # spaces
    for j in range(n - i):
        print(' ', end=' ')
    # stars
    for j in range(2 * i - 1):
        print('*', end=' ')
    print()

```

18.Left Aligned Pyramid.

```python
n = 5

for i in range(1, n + 1):
    for j in range(i):
        print('*', end=' ')
    print()
```

19.Right Aligned pyramid patterns.

```python
n = 5

for i in range(1, n + 1):
    
    for j in range(n - i):
        print(' ', end=' ')
    
    for j in range(i):
        print('*', end=' ')
    print()
```
